```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("jobs_in_data_science.csv")
columna1 = "salary"  # columna salary
columna2 = "salary_in_usd"  # columna salary_in_usd

salarios1 = df[columna1].dropna()  # Elimina valores nulos
salarios2 = df[columna2].dropna()  # Elimina valores nulos
plt.figure(figsize=(10, 6))
plt.hist(salarios1, bins=30, alpha=0.5, label="Salario", color="blue", edgecolor="black")
plt.hist(salarios2, bins=30, alpha=0.5, label="Salario en USD", color="red", edgecolor="black")

plt.xlabel("Salario")
plt.ylabel("Frecuencia")
plt.title("DistribuciÃ³n de Salarios")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("jobs_in_data_science.csv")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="experience_level", y="salary_in_usd", alpha=0.7, color="blue")

plt.xlabel("Nivel de Experiencia")
plt.ylabel("Salario en USD")
plt.title("RelaciÃ³n entre Nivel de Experiencia y Salario")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("jobs_in_data_science.csv")
correlation_matrix = df.corr(numeric_only=True)  # Columnas numÃ©ricas
print(correlation_matrix)
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matriz de CorrelaciÃ³n")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("jobs_in_data_science.csv")
df["salary_diff"] = df["salary"].diff()  # Diferencia entre filas consecutivas
mean_diff = df["salary_diff"].mean()  # Calcula el promedio de esas diferencias

print(f"Diferencia promedio de salario: {mean_diff:.2f}")
```
